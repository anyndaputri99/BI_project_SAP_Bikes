```python
# untuk membuat inisial nama yang dapat dimasukkan ke kolom name_initials
name_initial_list = []
for index, row in employess.iterrows():
    first_name_initial = row["NAME_FIRST"][0]
    last_name_initial = row["NAME_LAST"][0]
    if isinstance(row["NAME_MIDDLE"], float):  # untuk mengecualikan value float
        middle_name_initial = ""
    else:
        middle_name_initial = row["NAME_MIDDLE"][0]
    name_initial = f"{first_name_initial}{middle_name_initial}{last_name_initial}"
    name_initial_list.append(name_initial)
```

```python
employess["NAME_INITIALS"] = name_initial_list
```

```python
# membuat function untuk merubah format tanggal dari integer ke timestamp
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}-{month}-{day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
employess["validity_startdate"] = employess["validity_startdate"].apply(format_timestamp)
```
