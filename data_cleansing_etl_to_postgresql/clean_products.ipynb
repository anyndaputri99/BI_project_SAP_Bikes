```python
# mengubah format date string ke dalam bentuk timestamp
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}-{month}-{day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
products["created_at"] = products["created_at"].apply(format_timestamp)
```

```python
products["changed_at"] = products["changed_at"].apply(format_timestamp)
```

```python
# extract dari database
engine = create_engine("postgresql://postgres:260399@localhost:5432/postgres")
connection = engine.connect()
```

```python
queries = [
    """
    ALTER TABLE base_plat_sap_bikes.products_clean 
    ADD CONSTRAINT prodpartner_cons FOREIGN KEY(supplier_partner_id) 
    REFERENCES base_plat_sap_bikes.business_partners_clean(partner_id)
    """,
    """
    ALTER TABLE base_plat_sap_bikes.products_clean 
    ADD CONSTRAINT prodemploy_cons FOREIGN KEY(created_by) 
    REFERENCES base_plat_sap_bikes.employess_clean(employee_id)
    """,
    """
    ALTER TABLE base_plat_sap_bikes.products_clean 
    ADD CONSTRAINT prodcate_cons FOREIGN KEY(category_id) 
    REFERENCES base_plat_sap_bikes.product_category_clean(category_id)
    """
]

try:
    for query in queries:
        try:
            # Begin a new transaction
            connection.begin()
            # Execute the query
            connection.execute(text(query))
            # Commit the transaction
            connection.commit()
            print("Constraint added successfully.")
        except SQLAlchemyError as e:
            # Rollback in case of an error
            connection.rollback()
            print(f"Error executing query: {e}")
finally:
    # Close the connection
    connection.close()
```
