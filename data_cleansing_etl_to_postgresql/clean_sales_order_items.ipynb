```python
# membuat function untuk merubah format tanggal dari integer ke timestamp
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}-{month}-{day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
sales_order_items["delivery_date"] = sales_order_items["delivery_date"].apply(format_timestamp)
```

VALIDITY

```python
sales_order_items.describe([.25, .50, .75, .90, 1])
```

```python
sales_order_items[sales_order_items["delivery_date"] > pd.Timestamp('2020-01-01')] # terdapat data outliers pada delivery_date
# 3 total order_id (500000112, 500000164, 500000308)
# jika dijumlahkan data tersebut berdasarkan jumlah order dan net_amount, masih di bawah 5%, sehingga data bisa ditake out
```

```python
sales_order_items_clean = sales_order_items[sales_order_items["delivery_date"] < pd.Timestamp('2020-01-01')]
```

EXTRACT TO POSTGRESQL

```python
# extract dari database
engine = create_engine("postgresql://postgres:260399@localhost:5432/postgres")
connection = engine.connect()
```

```python
sales_order_items_clean.to_sql("order_details_clean",
                                 con=connection,
                                 schema="base_plat_sap_bikes",
                                 if_exists="replace",
                                 index=True,
                                 method="multi"
)
```

```python
# setting up the primary key
query_pk = """
alter table base_plat_sap_bikes.order_details_clean
add primary key(index)
"""
result = connection.execute(text(query_pk))
connection.commit()
connection.close()
```
