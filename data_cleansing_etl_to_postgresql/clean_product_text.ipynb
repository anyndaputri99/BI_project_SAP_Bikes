EXTRACT TO POSTGRESQL

```python
# extract dari database
engine = create_engine("postgresql://postgres:260399@localhost:5432/postgres")
connection = engine.connect()
```

```python
product_text_clean.to_sql("product_text_clean",
                                 con=connection,
                                 schema="base_plat_sap_bikes",
                                 if_exists="replace",
                                 index=True,
                                 method="multi"
)
```

```python
# setting up the primary key
query_pk = """
alter table base_plat_sap_bikes.product_text_clean
add primary key(index, product_id)
"""
result = connection.execute(text(query_pk))
connection.commit()
connection.close()
```
