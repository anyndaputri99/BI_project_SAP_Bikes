```python
# mengubah format ke timestamp dari string
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}-{month}-{day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
sales_orders["created_at"] = sales_orders["created_at"].apply(format_timestamp)
```

```python
sales_orders["changed_at"] = sales_orders["changed_at"].apply(format_timestamp)
```

```python
sales_orders
```

```python
# mengubah format pada kolom fiscal_year_period dari integer menjadi string date -> lebih readable dalam format yyyymm
def format_date(date):
    fiscal_string = str(date)
    year, month = fiscal_string[:4], fiscal_string[5:]
    formatted_date = f"{year}{month}"
    datetime_format = pd.to_datetime(formatted_date, format='%Y%m')
    return datetime_format.strftime('%Y%m')
```

```python
sales_orders["fiscal_year_period"] = sales_orders["fiscal_year_period"].apply(format_date)
```
