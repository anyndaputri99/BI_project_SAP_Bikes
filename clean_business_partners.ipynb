```python
# function untuk mengubah format ke timestamp
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}, {month}, {day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
business_partners_updated["created_at"] = business_partners_updated["created_at"].apply(format_timestamp)
```

```python
business_partners_updated["changed_at"] = business_partners_updated["changed_at"].apply(format_timestamp)
```

```python
# extract dari database
engine = create_engine("postgresql://postgres:260399@localhost:5432/postgres")
connection = engine.connect()
```

```python
queries = [
    """
    ALTER TABLE base_plat_sap_bikes.business_partners_clean 
    ADD CONSTRAINT partaddress_cons FOREIGN KEY(address_id) 
    REFERENCES base_plat_sap_bikes.addresses_clean(address_id)
    """,
    """
    ALTER TABLE base_plat_sap_bikes.business_partners_clean 
    ADD CONSTRAINT partemploy_cons FOREIGN KEY(created_by) 
    REFERENCES base_plat_sap_bikes.employess_clean(employee_id)
    """
]

try:
    for query in queries:
        connection.execute(text(query))
    
    # Commit the transaction
    connection.commit()
    print("Constraints added successfully.")
except SQLAlchemyError as e:
    # Rollback in case of an error
    connection.rollback()
    print(f"Error: {e}")
finally:
    # Close the connection
    connection.close()
```
