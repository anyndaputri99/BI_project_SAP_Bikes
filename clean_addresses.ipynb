```python
# membuat function untuk merubah format tanggal dari integer ke timestamp
def format_timestamp(date):
    string_date = str(date)
    year, month, day = string_date[0:4], string_date[4:6], string_date[6:8]
    date_format = f"{year}-{month}-{day}"
    timestamp = pd.Timestamp(date_format)
    return timestamp
```

```python
addresses_updated["validity_startdate"] = addresses_updated["validity_startdate"].apply(format_timestamp)
```

```python
addresses_updated[["address_id", "validity_enddate"]].groupby(by=["validity_enddate"]).nunique() # penulisan tanggal tidak dalam bentuk format timestamp
```

```python
addresses_updated["validity_enddate"] = "20501231"
```

```python
addresses_updated["validity_enddate"] = addresses_updated["validity_enddate"].apply(format_timestamp)
```

```python
addresses_updated.info() # address_id, address_type, latitude, dan longitude masih dalam bentuk integer
```

```python
addresses_updated["address_id"] = addresses_updated["address_id"].astype(str)
```

```python
addresses_updated["address_type"] = addresses_updated["address_type"].astype(str)
```

```python
addresses_updated["latitude"] = addresses_updated["latitude"].astype(str)
```

```python
addresses_updated["longitude"] = addresses_updated["longitude"].astype(str)
```
